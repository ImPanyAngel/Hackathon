```python
import numpy as np
import pandas as pd
import pickle
import os
import sqlite3
```

```python
conn =  sqlite3.connect(r"Sqlite3.db")
try:
    df  = pd.read_sql("SELECT * FROM posts",conn)
except sqlite3.Error as error:
    print(error)
```

```python
df.head()
```

```python
posts_df = df[['post_description','course','skills']]
posts_df
```

```python
posts_df['overview'] = posts_df['post_description']+' ' +posts_df['course']+' ' +posts_df['skills']
posts_df
```

```python
posts_df = posts_df.drop_duplicates()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features=3000,stop_words='english')
X = vectorizer.fit_transform(posts_df['overview']).toarray()
```

```python
vectorizer.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_mat = cosine_similarity(X)
```

```python
d = pd.DataFrame(similarity_mat)
d.head()
```

```python
sorted(list(enumerate(similarity_mat[0])),reverse=True,key=lambda x:x[1])[1:10]
```

```python
def post_recommend(course):
    postIndex  = df[df['course']==course].index[0]
    distances_ = similarity_mat[postIndex]
    post_list = sorted(list(enumerate(distances_)),reverse=True,key=lambda x:x[1])[1:10]
    recommended_df = df.iloc[[i[0] for i in post_list]]
    return recommended_df
```

```python
post_recommend("Biology")
```

```python
directory_path = ""
with open(os.path.join(directory_path, "postList.pkl"), 'wb') as f:
    pickle.dump(df.to_dict(), f)

with open(os.path.join(directory_path, "similarity.pkl"), 'wb') as f:
    pickle.dump(similarity_mat, f)
```
